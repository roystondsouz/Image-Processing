```python
import matplotlib.pyplot as plt
def show_image_contour(image, contours):
    plt.figure()
    for n, contour in enumerate(contours):
        plt.plot(contour[:,1],contour[:,0], linewidth=3)
    plt.imshow(image,interpolation='nearest',cmap='gray_r')
    plt.title('contours')
    plt.axis('off')
  
```

```python
from skimage import measure, data
    
#obtain the horse image
kidney_image=data.kidney()
    
#find the contours with a constant level value of 0.8
contours= measure.find_contours(kidney_image, level=0.8)
    
#show the image with contours found
show_image_contour(kidney_image, contours)
```

```python
from skimage.io import imread
from skimage.filters import threshold_otsu
from skimage import color

image_dices = imread('diceimg.png')

#make image grayscale
image_dices = color.rgb2gray(image_dices)

#obtain optical threshold value
thresh = threshold_otsu(image_dices)

#Apply thresholding
binary = image_dices > thresh

#find contours at a constant value of 0.8
contours = measure.find_contours(binary,level=0.8)

#SHOW IMAGE
show_image_contour(image_dices,contours)
```

```python
import numpy as np
#create  list with the shaPE of each conours
shape_contours=[cnt.shape[0] for cnt in contours]

#set 50 as amx  sizw of the . shape
max_dots_shape=50

#count .s in contours excluding bigger than . size
dots_contours=[cnt for cnt in contours if np.shape(cnt)[0] < max_dots_shape]

#show all contors found
show_image_contour(binary,contours)

#print the dice no.
print('Dice`s dots number:{}.'.format(len(dots_contours)))
```
